```python
from sklearn.feature_selection import f_classif,VarianceThreshold,SelectFromModel,RFE
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

data=pd.read_csv('/Users/hurrywish/Downloads/Python_sklearn/【第01部分 核心教程】sklearn深度学习框架入门与实战 3套/02.2019年Scikit-learn的机器学习入门到精通（附讲义、代码、数据集、参考资料） 258课/00.配套资料（代码、讲义、数据等）/00.配套资料（代码、讲义、数据等）/03数据预处理和特征工程/digit recognizor.csv')


sample=data[data.columns[1:]]
label=data[data.columns[0]]

clfs=dict()
clfs['DTC']=DecisionTreeClassifier(random_state=1)
clfs['RFC']=RandomForestClassifier(random_state=1,n_jobs=-1)
#clfs['KNC']=KNeighborsClassifier(n_jobs=-1)

samples=dict()

samples['sample_0']=sample
samples['samples_median_V']=VarianceThreshold(np.median(sample.var().values)).fit_transform(sample)
samples['samples_mean_V']=VarianceThreshold(np.mean(sample.var().values)).fit_transform(sample)
samples['samples_mean_Emb']=SelectFromModel(RandomForestClassifier(),threshold=0.005).fit_transform(sample,label)
samples['samples_RFE']=RFE(estimator=RFC,n_features_to_select=340,step=50).fit_transform(sample,label)

for i in clfs:
    clf=clfs[i]
    for j in samples:
        vals=samples[j]
        score_cross=cross_val_score(clf,vals,label,cv=5).mean()
        print('clf:',i)
        print('sample:',j)
        print('score:',score_cross)
```

```python
sample=data[data.columns[1:]]
label=data[data.columns[0]]

RFC=RandomForestClassifier(n_jobs=-1)
RFC.fit(sample,label)
high_limit=max(RFC.feature_importances_)
threshold_range=np.linspace(0,high_limit,50)

scores=list()
for i in threshold_range:
    sample_emb=SelectFromModel(RFC,threshold=i).fit_transform(sample,label)
    score=cross_val_score(RFC,sample,label,cv=5).mean()
    scores.append(score)

print(scores)
plt.plot(threshold_range,scores)
```
